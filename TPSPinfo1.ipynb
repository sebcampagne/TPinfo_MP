```python
import numpy as np, matplotlib.pyplot as plt , numpy.random as rd
import scipy.integrate as integr, scipy.optimize as resol
```

```python
def f(x):
    return integr.quad(lambda t:abs(t-x),0,1)[0]
```

```python
tx=np.linspace(-1,2,1000)
ty=[f(a) for a in tx ]
plt.grid(),plt.plot(tx, ty)
plt.show()
```

```python
def h(x,t):
    return -x/(1+t) + np.sin(t)
```

```python
tx=np.linspace(-0.9,8,100)
x=np.linspace(-10,10,11)
for x0 in x:
    y=integr.odeint(h,x0,tx)
    plt.plot(tx,y)
plt.show()
```

```python
def g(x,t):
    return 2*x/t -t*np.cos(1/t)
T0=np.linspace(-1/np.pi , -0.2,10)
for t0 in T0:
    tx=np.linspace(t0,-10**(-4),100)
    y=integr.odeint (g,0,tx)
    plt.plot(tx,y)
plt.show()
```

```python
for n in range(100,1001):
    tx=np.linspace(0.001,0.1,n)
    y=integr.odeint (g,0,tx)
    plt.plot(tx,y)
plt.show()
```
